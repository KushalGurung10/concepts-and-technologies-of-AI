3. Missing Values & Data Cleaning:

```python
print(hdi_2022_df.isnull().sum())

```

```python
hdi_2022_df = hdi_2022_df.drop_duplicates()


hdi_2022_df['hdi'] = pd.to_numeric(hdi_2022_df['hdi'], errors='coerce')
hdi_2022_df['gross_inc_percap'] = pd.to_numeric(hdi_2022_df['gross_inc_percap'], errors='coerce')


hdi_2022_df = hdi_2022_df.dropna(subset=['hdi', 'gross_inc_percap'])
```

4. Basic Statistics:

```python
mean_hdi = hdi_2022_df['hdi'].mean()
median_hdi = hdi_2022_df['hdi'].median()
std_hdi = hdi_2022_df['hdi'].std()

print("Mean HDI:", mean_hdi)
print("Median HDI:", median_hdi)
print("Standard Deviation:", std_hdi)
```

```python
print("Highest HDI Country:", hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax(), 'country'])
print("Lowest HDI Country:", hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(), 'country'])
```

5. Filtering and Sorting:

```python
high_hdi_df = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]
```

```python
high_hdi_sorted = high_hdi_df.sort_values(
    by='gross_inc_percap',
    ascending=False
)
```

```python
high_hdi_sorted[['country', 'hdi', 'gross_inc_percap']].head(10)
```

6. Adding HDI Category Column:
HDI Categories:

Low < 0.550

Medium 0.550 – 0.699

High 0.700 – 0.799

Very High ≥ 0.800

```python
bins = [0, 0.550, 0.699, 0.799, 1.0]
labels = ['Low', 'Medium', 'High', 'Very High']

hdi_2022_df['HDI_Category'] = pd.cut(
    hdi_2022_df['hdi'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

print(hdi_2022_df[['country', 'hdi', 'HDI_Category']].head(15))

hdi_2022_df.to_csv('HDI_category_added.csv', index=False)
```

4 Problem 2

Advanced HDI Exploration

1. Create South Asia Subset:

```python
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]
sa_df = df[df['country'].isin(south_asia_countries)]
sa_df.to_csv('/content/drive/MyDrive/AI/HDI_SouthAsia.cs', index=False)

```

2. Composite Development Score:

```python
sa_df = sa_df.copy()
sa_df['Composite Score'] = 0.3 * sa_df['life_expectancy'] + 0.3 * sa_df['gross_inc_percap']
sa_top5 = sa_df.sort_values('Composite Score', ascending=False).head(5)


plt.figure(figsize=(6,4))
plt.barh(sa_top5['country'], sa_top5['Composite Score'])
plt.xlabel('Composite Score')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.show()
```

3. Outlier Detection:

```python
Q1 = sa_df['hdi'].quantile(0.25)
Q3 = sa_df['hdi'].quantile(0.75)
IQR = Q3 - Q1
outliers_hdi = sa_df[(sa_df['hdi'] < Q1 - 1.5*IQR) | (sa_df['hdi'] > Q3 + 1.5*IQR)]

plt.scatter(sa_df['gross_inc_percap'], sa_df['hdi'], color='blue')
plt.scatter(outliers_hdi['gross_inc_percap'], outliers_hdi['hdi'], color='red')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('HDI vs GNI per Capita (Outliers in Red)')
plt.show()
```

4. Exploring Metric Relationships:

```python
from scipy.stats import pearsonr


if 'gender_development' in sa_df.columns:

    filtered_df = sa_df.dropna(subset=['gender_development', 'hdi'])
    if not filtered_df.empty:
        corr, _ = pearsonr(filtered_df['gender_development'], filtered_df['hdi'])
        print("Pearson correlation (Gender Development vs HDI):", corr)
    else:
        print("Cannot compute correlation: insufficient non-null data after filtering.")
else:
    print("'gender_development' column not found in DataFrame.")
```

5. Gap Analysis:

```python
sa_df = sa_df.copy()
sa_df['GNI_HDI_Gap'] = sa_df['gross_inc_percap'] - sa_df['hdi']
gap_top3 = sa_df.sort_values('GNI_HDI_Gap', ascending=False).head(3)
gap_bottom3 = sa_df.sort_values('GNI_HDI_Gap').head(3)

print("\nTop 3 GNI-HDI Gap Countries (Positive):\n", gap_top3[['country','GNI_HDI_Gap']])
print("\nTop 3 GNI-HDI Gap Countries (Negative):\n", gap_bottom3[['country','GNI_HDI_Gap']])
```

5 Problem 3

Comparative Regional Analysis: South Asia vs Middle East

1. Create Middle East Subset:

```python
middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
                         "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
                         "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]


sa_2020_2022 = hdi_2020_2022_df[hdi_2020_2022_df['country'].isin(south_asia_countries)]
me_2020_2022 = hdi_2020_2022_df[hdi_2020_2022_df['country'].isin(middle_east_countries)]

sa_2020_2022.to_csv("/content/drive/MyDrive/AI/HDI_SouthAsia_2020_2022.csv", index=False)
me_2020_2022.to_csv("/content/drive/MyDrive/AI/HDI_MiddleEast_2020_2022.csv", index=False)
```

2. Descriptive Statistics



```python
print("Descriptive Statistics:")
print("South Asia HDI Mean:", sa_2020_2022['hdi'].mean(),
      "Standard Deviation:", sa_2020_2022['hdi'].std())
print("Middle East HDI Mean:", me_2020_2022['hdi'].mean(),
      "Standard Deviation:", me_2020_2022['hdi'].std())
```

and 3. Top and Bottom Performers:


```python
sa_top3 = sa_2020_2022.groupby('country')['hdi'].mean().sort_values(ascending=False).head(3)
sa_bottom3 = sa_2020_2022.groupby('country')['hdi'].mean().sort_values().head(3)

me_top3 = me_2020_2022.groupby('country')['hdi'].mean().sort_values(ascending=False).head(3)
me_bottom3 = me_2020_2022.groupby('country')['hdi'].mean().sort_values().head(3)

print("\nSouth Asia Top 3 Countries by HDI:\n", sa_top3)
print("\nSouth Asia Bottom 3 Countries by HDI:\n", sa_bottom3)
print("\nMiddle East Top 3 Countries by HDI:\n", me_top3)
print("\nMiddle East Bottom 3 Countries by HDI:\n", me_bottom3)
```

4. Metric Comparisons:

```python
metrics = ['life_expectancy', 'gross_inc_percap', 'gender_development']

south_means = sa_2020_2022[metrics].mean()
middle_means = me_2020_2022[metrics].mean()


comparison_df = pd.DataFrame({
    'South Asia': south_means,
    'Middle East': middle_means
})

comparison_df.plot(kind='bar', figsize=(8,5))
plt.title("Metric Comparison: South Asia vs Middle East (2020-2022)")
plt.ylabel("Average Value")
plt.xlabel("Metric")
plt.xticks(rotation=0)
plt.show()
```

5. HDI Disparity:


```python
south_range = sa_2020_2022['hdi'].max() - sa_2020_2022['hdi'].min()
middle_range = me_2020_2022['hdi'].max() - me_2020_2022['hdi'].min()

south_cv = sa_2020_2022['hdi'].std() / sa_2020_2022['hdi'].mean()
middle_cv = me_2020_2022['hdi'].std() / me_2020_2022['hdi'].mean()

print(f"South Asia HDI Range: {south_range:.3f}")
print(f"Middle East HDI Range: {middle_range:.3f}")
print(f"South Asia HDI Coefficient of Variation: {south_cv:.3f}")
print(f"Middle East HDI Coefficient of Variation: {middle_cv:.3f}")

plt.figure(figsize=(8,5))
sns.boxplot(data=[sa_2020_2022['hdi'], me_2020_2022['hdi']],
            palette=["skyblue", "salmon"])
plt.xticks([0,1], ['South Asia', 'Middle East'])
plt.ylabel("HDI")
plt.title("HDI Distribution: South Asia vs Middle East (2020-2022)")
plt.show()
```

6. Correlation Analysis:

```python
south_corr_life = sa_2020_2022[['hdi', 'life_expectancy']].corr()
middle_corr_life = me_2020_2022[['hdi', 'life_expectancy']].corr()

south_corr_gni = sa_2020_2022[['hdi', 'gross_inc_percap']].corr()
middle_corr_gni = me_2020_2022[['hdi', 'gross_inc_percap']].corr()

print("\nSouth Asia HDI vs Life Expectancy Correlation:\n", south_corr_life)
print("\nMiddle East HDI vs Life Expectancy Correlation:\n", middle_corr_life)

print("\nSouth Asia HDI vs GNI per Capita Correlation:\n", south_corr_gni)
print("\nMiddle East HDI vs GNI per Capita Correlation:\n", middle_corr_gni)
```

7. Outlier Detection:

```python
Q1_south = sa_2020_2022['hdi'].quantile(0.25)
Q3_south = sa_2020_2022['hdi'].quantile(0.75)
IQR_south = Q3_south - Q1_south

outliers_south = sa_2020_2022[
    (sa_2020_2022['hdi'] < Q1_south - 1.5 * IQR_south) |
    (sa_2020_2022['hdi'] > Q3_south + 1.5 * IQR_south)
]

print("\nSouth Asia HDI Outliers:\n", outliers_south[['country', 'year', 'hdi']])


Q1_middle = me_2020_2022['hdi'].quantile(0.25)
Q3_middle = me_2020_2022['hdi'].quantile(0.75)
IQR_middle = Q3_middle - Q1_middle

outliers_middle = me_2020_2022[
    (me_2020_2022['hdi'] < Q1_middle - 1.5 * IQR_middle) |
    (me_2020_2022['hdi'] > Q3_middle + 1.5 * IQR_middle)
]

print("\nMiddle East HDI Outliers:\n", outliers_middle[['country', 'year', 'hdi']])



plt.figure(figsize=(8,5))

plt.scatter(sa_2020_2022['gross_inc_percap'], sa_2020_2022['hdi'], color='blue', label='South Asia')
plt.scatter(outliers_south['gross_inc_percap'], outliers_south['hdi'], color='red', label='South Asia Outliers', marker='x')

plt.scatter(me_2020_2022['gross_inc_percap'], me_2020_2022['hdi'], color='green', label='Middle East')
plt.scatter(outliers_middle['gross_inc_percap'], outliers_middle['hdi'], color='orange', label='Middle East Outliers', marker='x')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('HDI vs GNI per Capita with Outliers Highlighted')
plt.legend()
plt.show()
```
